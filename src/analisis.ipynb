### Caracteristicas de las variables

```python
df_final = mf.bajar_fichero(mf.crear_cliente(), path_server=f"grupo3/raw/Final/prueba.parquet", type="df")
print("*"*20)
print('Se presenta el analisis general del dataset final en los siguientes puntos:')
print("*"*20)
print('Representación de las primeras filas del dataset:')
print(df_final.head(10))
print(f"La longitud total de los datos es de: {len(df_final)}")
```

```python
print('Información del tipo de variables del dataset:')
df_final.info()
```

```python
print('Conteo de valores no-nulos por columna:')
print(df_final.notna().sum())
print(f'Hay {df_final.notna().sum().sum()} valores no-nulos de un total de {df_final.size} celdas, lo que representa un {df_final.notna().sum().sum()/df_final.size*100:.2f}% de los datos.')
```

```python
print('Presentacion de informacion estadistica de las variables:')
df_final.describe()
```

```python
print('Frecuencia de las distintas variables categóricas:')
for col in df_final.select_dtypes(include=['object']).columns:
    print('*'*20+f'Columna: {col}'+'*'*20)
    # print(df_final[col].value_counts(dropna=False))
    print('Frecuencia relativa:')
    print(df_final[col].value_counts(dropna=False, normalize=True))
    print('Numero de valores únicos:', df_final[col].nunique())
```
